```python
import requests
import io
import fitz  # PyMuPDF for PDF extraction
from docx import Document  # For DOCX extraction
import spacy
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# Ensure necessary NLTK resources are downloaded
nltk.download('stopwords')
nltk.download('punkt')

# Load the transformer-based SpaCy model
nlp = spacy.load("en_core_web_trf")

# Define job titles and associated technologies
job_roles = [
    {
        "job_title": "Software Developer / Software Engineer",
        "technology_stack": {
            "Programming Languages": ["Java", "Python", "C#", "C++", "JavaScript", "Ruby"],
            "Frameworks": ["Spring Boot", "Django", "ASP.NET", "Ruby on Rails"],
            "Databases": ["MySQL", "PostgreSQL", "MongoDB", "Oracle"],
            "Version Control": ["Git", "SVN"],
            "CI/CD Tools": ["Jenkins", "GitLab CI", "CircleCI"]
        }
    },
    {
        "job_title": "Frontend Developer",
        "technology_stack": {
            "Languages": ["HTML", "CSS", "JavaScript", "TypeScript"],
            "Frameworks/Libraries": ["React.js", "Angular", "Vue.js", "Svelte"],
            "Tools": ["Webpack", "Babel", "npm", "Yarn"],
            "UI/UX Tools": ["Figma", "Adobe XD", "Sketch"]
        }
    },
    {
        "job_title": "Backend Developer",
        "technology_stack": {
            "Languages": ["Java", "Python", "Ruby", "PHP", "Node.js"],
            "Frameworks": ["Express.js", "Flask", "Django", "Laravel", "Spring"],
            "Databases": ["MySQL", "PostgreSQL", "MongoDB", "Redis", "Elasticsearch"],
            "API Tools": ["RESTful", "GraphQL", "gRPC"]
        }
    },
    {
        "job_title": "Full Stack Developer",
        "technology_stack": {
            "Combination of Frontend and Backend stacks": [
                "HTML", "CSS", "JavaScript", "React.js", "Angular", "Vue.js", "Spring Boot", "Django", "Express.js", "Flask"
            ],
            "DevOps Tools": ["Docker", "Kubernetes", "Jenkins", "AWS", "Azure"],
            "Databases": ["SQL", "NoSQL"]
        }
    },
    {
        "job_title": "DevOps Engineer",
        "technology_stack": {
            "CI/CD Tools": ["Jenkins", "Travis CI", "GitLab CI/CD", "CircleCI"],
            "Containerization": ["Docker", "Kubernetes"],
            "Cloud Platforms": ["AWS", "Azure", "Google Cloud"],
            "Scripting": ["Python", "Bash", "Perl"],
            "Monitoring Tools": ["Prometheus", "Grafana", "ELK Stack"]
        }
    },
    {
        "job_title": "Data Engineer",
        "technology_stack": {
            "Programming Languages": ["Python", "Java", "Scala"],
            "Big Data Tools": ["Apache Hadoop", "Spark", "Kafka"],
            "ETL Tools": ["Apache NiFi", "Talend", "Informatica"],
            "Databases": ["SQL", "NoSQL (Cassandra, MongoDB)", "Data Warehouses (Redshift, Snowflake)"],
            "Cloud Platforms": ["AWS", "Azure", "Google Cloud"]
        }
    },
    {
        "job_title": "Data Scientist",
        "technology_stack": {
            "Programming Languages": ["Python", "R", "SQL"],
            "Libraries/Tools": ["Pandas", "NumPy", "Scikit-learn", "TensorFlow", "PyTorch"],
            "Data Visualization": ["Matplotlib", "Seaborn", "Power BI", "Tableau"],
            "Databases": ["MySQL", "PostgreSQL", "MongoDB", "Spark"]
        }
    },
    {
        "job_title": "Machine Learning Engineer",
        "technology_stack": {
            "Programming Languages": ["Python", "R", "Java", "Scala"],
            "Frameworks": ["TensorFlow", "PyTorch", "Scikit-learn", "Keras"],
            "Tools": ["Jupyter", "Apache Spark", "Hadoop"],
            "Cloud Platforms": ["AWS Sagemaker", "Azure ML", "Google AI Platform"]
        }
    },
    {
        "job_title": "Cloud Architect",
        "technology_stack": {
            "Cloud Platforms": ["AWS", "Azure", "Google Cloud", "IBM Cloud"],
            "Containerization": ["Docker", "Kubernetes"],
            "Infrastructure as Code": ["Terraform", "CloudFormation"],
            "Security Tools": ["IAM", "Cloud Security", "VPN", "Firewalls"]
        }
    },
    {
        "job_title": "Cybersecurity Analyst/Engineer",
        "technology_stack": {
            "Security Tools": ["Firewalls", "VPNs", "IDS/IPS", "SIEM (Splunk, QRadar)"],
            "Programming": ["Python", "Bash", "PowerShell"],
            "Ethical Hacking": ["Kali Linux", "Metasploit"],
            "Encryption": ["SSL/TLS", "RSA", "AES"],
            "Cloud Security": ["AWS Security Hub", "Azure Security Center"]
        }
    },
    {
        "job_title": "Database Administrator (DBA)",
        "technology_stack": {
            "Databases": ["Oracle", "MySQL", "PostgreSQL", "SQL Server", "MongoDB"],
            "Tools": ["SQL", "PL/SQL", "Shell Scripting", "Backup & Recovery tools"],
            "Monitoring": ["Nagios", "SolarWinds", "Zabbix"],
            "Cloud Services": ["AWS RDS", "Azure SQL Database"]
        }
    },
    {
        "job_title": "System Administrator",
        "technology_stack": {
            "Operating Systems": ["Linux (RedHat, Ubuntu)", "Windows Server"],
            "Scripting": ["Bash", "PowerShell", "Python"],
            "Virtualization": ["VMware", "Hyper-V"],
            "Networking": ["TCP/IP", "DNS", "DHCP"],
            "Monitoring": ["Nagios", "Zabbix", "Prometheus"]
        }
    },
    {
        "job_title": "QA Engineer / Test Automation Engineer",
        "technology_stack": {
            "Testing Tools": ["Selenium", "JUnit", "TestNG", "Cypress", "QTP"],
            "CI/CD": ["Jenkins", "Bamboo", "GitLab CI"],
            "Scripting": ["Python", "Java", "JavaScript"],
            "Test Management": ["JIRA", "TestRail", "Zephyr"],
            "Performance Testing": ["JMeter", "LoadRunner"]
        }
    },
    {
        "job_title": "UI/UX Designer",
        "technology_stack": {
            "Design Tools": ["Adobe XD", "Figma", "Sketch", "InVision"],
            "Prototyping": ["Marvel", "Axure"],
            "Frontend Skills": ["HTML", "CSS", "JavaScript (optional)"],
            "Graphic Design": ["Adobe Photoshop", "Illustrator"]
        }
    },
    {
        "job_title": "Product Manager",
        "technology_stack": {
            "Project Management Tools": ["Jira", "Trello", "Asana", "Monday.com"],
            "Prototyping": ["Figma", "InVision"],
            "Analytics": ["Google Analytics", "Mixpanel", "Tableau"],
            "Communication Tools": ["Slack", "Microsoft Teams", "Zoom"]
        }
    },
    {
        "job_title": "Business Analyst",
        "technology_stack": {
            "Requirements Tools": ["Jira", "Confluence", "Trello"],
            "Data Analysis": ["Excel", "Power BI", "Tableau"],
            "SQL": ["MySQL", "SQL Server", "PostgreSQL"],
            "Process Modeling": ["Visio", "Lucidchart", "Balsamiq"]
        }
    },
    {
        "job_title": "Site Reliability Engineer (SRE)",
        "technology_stack": {
            "Monitoring": ["Prometheus", "Grafana", "Nagios"],
            "CI/CD Tools": ["Jenkins", "GitLab CI", "Travis CI"],
            "Cloud Platforms": ["AWS", "GCP", "Azure"],
            "Scripting": ["Python", "Bash", "Go"],
            "Infrastructure as Code": ["Terraform", "Ansible", "Puppet"]
        }
    },
    {
        "job_title": "Network Engineer",
        "technology_stack": {
            "Networking": ["Cisco", "Juniper", "Huawei", "Palo Alto Networks"],
            "Protocols": ["TCP/IP", "BGP", "OSPF", "MPLS"],
            "Tools": ["Wireshark", "SolarWinds", "Nagios"],
            "Security": ["Firewalls", "VPNs", "IDS/IPS"]
        }
    },
    {
        "job_title": "Technical Support Engineer",
        "technology_stack": {
            "Operating Systems": ["Windows", "Linux", "macOS"],
            "Remote Support": ["TeamViewer", "AnyDesk"],
            "Ticketing Systems": ["Zendesk", "Jira Service Desk"],
            "Networking": ["TCP/IP", "VPN", "DNS"]
        }
    }

]

# Define additional entities for extraction
additional_entities = {
    "WORK_OF_ART": ["art", "painting", "sculpture", "music", "film", "literature"],
    "EXPERIENCE": ["experience", "project", "work", "internship", "role", "position"]
}

# Function to return job title and technology stack
def get_job_technologies(job_title):
    for role in job_roles:
        if role["job_title"] == job_title:
            return role["job_title"], role["technology_stack"]
    return None, None

# Function to handle user login and fetch resume
def get_user_resume(email):
    resume_url = get_resume_url(email)

    if resume_url:
        print(f"Resume URL: {resume_url}")
        resume_text = extract_resume_text(resume_url)

        # Preprocess the resume text
        preprocessed_text = preprocess_text(resume_text)

        # Extract and classify entities
        entities = extract_and_classify_entities(preprocessed_text)

        # Save the preprocessed text and entities
        save_preprocessed_text(preprocessed_text)
        save_entities(entities)

        print("Processing complete!")
    else:
        print("Failed to fetch the resume URL.")

# Fetch the resume URL
def get_resume_url(email):
    url = f"https://aitq-api-t4yr74gyka-uc.a.run.app/users/{email}/resume"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json().get("resume_url")
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None

# Extract text from PDF
def extract_text_from_pdf(url):
    response = requests.get(url)
    pdf_data = io.BytesIO(response.content)
    text = ""
    with fitz.open(stream=pdf_data, filetype="pdf") as pdf_doc:
        for page in pdf_doc:
            text += page.get_text()
    return text

# Extract text from DOCX
def extract_text_from_docx(url):
    response = requests.get(url)
    doc_data = io.BytesIO(response.content)
    document = Document(doc_data)
    text = '\n'.join([para.text for para in document.paragraphs])
    return text

# Extract and process text based on file type
def extract_resume_text(resume_url):
    if resume_url.endswith('.pdf'):
        return extract_text_from_pdf(resume_url)
    elif resume_url.endswith('.docx'):
        return extract_text_from_docx(resume_url)
    else:
        print("Unsupported file type.")
        return ""

# Preprocessing function
def preprocess_text(text):
    # Tokenization
    sentences = sent_tokenize(text)
    words = word_tokenize(text)

    # Normalization
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Stop Words Removal
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Lemmatization
    lemmatized_words = []
    for word in filtered_words:
        doc = nlp(word)
        lemmatized_words.append(doc[0].lemma_)

    # Combine all preprocessed text into a single string
    preprocessed_text = f"Sentences:\n{sentences}\n\nFiltered Words:\n{filtered_words}\n\nLemmatized Words:\n{lemmatized_words}"
    return preprocessed_text

# Save preprocessed text to a file
def save_preprocessed_text(preprocessed_text):
    output_preprocessed_path = "/content/preprocessed_text.txt"
    with open(output_preprocessed_path, 'w', encoding='utf-8') as file:
        file.write(preprocessed_text)
    print(f"Preprocessed text saved to '{output_preprocessed_path}'.")

# Function for Named Entity Recognition and Classification
def extract_and_classify_entities(text):
    doc = nlp(text)
    skills = get_skills_set()
    entities = []
    job_titles = []
    technologies = []
    work_of_art = []
    experience = []

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            entities.append((ent.text, "PERSON"))
        elif ent.label_ == "ORG":
            entities.append((ent.text, "ORG"))
        elif ent.label_ == "DATE":
            entities.append((ent.text, "DATE"))
        elif ent.label_ in {"NORP", "GPE", "LOC"}:
            entities.append((ent.text, "LOCATION"))
        elif ent.text in skills:
            entities.append((ent.text, "SKILL"))
        elif any(word in ent.text.lower() for word in additional_entities["WORK_OF_ART"]):
            work_of_art.append(ent.text)
        elif any(word in ent.text.lower() for word in additional_entities["EXPERIENCE"]):
            experience.append(ent.text)
        else:
            entities.append((ent.text, ent.label_))

    # Extract job titles and technologies based on context
    for sentence in doc.sents:
        sentence_text = sentence.text.lower()
        for role in job_roles:
            if any(title.lower() in sentence_text for title in [role["job_title"].lower()]):
                job_titles.append(role["job_title"])
                technologies.extend([tech for tech in role["technology_stack"].values()])

    return entities, job_titles, technologies, work_of_art, experience

# Save extracted entities to a file
def save_entities(entities):
    output_entity_path = "/content/extracted_entities.txt"
    with open(output_entity_path, 'w') as output_file:
        output_file.write("Extracted Entities:\n")
        for entity, label in entities[0]:
            output_file.write(f"{entity}: {label}\n")
        output_file.write("\nJob Titles:\n")
        for title in entities[1]:
            output_file.write(f"{title}\n")
        output_file.write("\nTechnologies:\n")
        for tech in set(entities[2]):
            output_file.write(f"{tech}\n")
        output_file.write("\nWorks of Art:\n")
        for work in entities[3]:
            output_file.write(f"{work}\n")
        output_file.write("\nExperience:\n")
        for exp in entities[4]:
            output_file.write(f"{exp}\n")
    print(f"Extracted and classified entities saved to '{output_entity_path}'.")

# Function to provide skills set
def get_skills_set():
    return {
        # Existing skills...
        "Python",
    "Java",
    "C++",
    "HTML",
    "CSS",
    "JavaScript",
    "SQL",
    "NoSQL",
    "TensorFlow",
    "PyTorch",
    "R",
    "Pandas",
    "AWS",
    "Azure",
    "AutoCAD",
    "SolidWorks",
    "SPICE",
    "MS Project",
    "Lean Manufacturing",
    "Excel",
    "QuickBooks",
    "SEO",
    "PPC",
    "Agile",
    "Scrum",
    "Tableau",
    "Google Analytics",
    "Medical Terminology",
    "Patient Care Skills",
    "EMR Systems",
    "Clinical Trials",
    "Photoshop",
    "Illustrator",
    "Writing and Editing",
    "Social Media Management",
    "Public Speaking",
    "PCR",
    "Spectroscopy",
    "Data Collection and Analysis",
    "Research Methodologies",
    "Taxation",
    "Auditing",
    "Financial Reporting",
    "Management Accounting",
    "Strategic Management",
    "Business Analytics",
    "Software Development",
    "Database Management",
    "Web Development",
    "Data Structures",
    "Algorithms",
    "Software Engineering",
    "Networking",

    # General Academic Skills
    "Critical Thinking",
    "Problem Solving",
    "Research Skills",
    "Analytical Skills",
    "Communication Skills",
    "Time Management",
    "Teamwork",
    "Adaptability",
    "Creativity",
    "Organization Skills",
    "Information Literacy",
    "Study Skills",
    "Attention to Detail",

    # Science and Technology Skills
    "Scientific Method",
    "Statistical Analysis",
    "Computational Thinking",

    # Mathematics Skills
    "Algebra",
    "Geometry",
    "Calculus",
    "Probability",

    # Social Sciences Skills
    "Qualitative Research",
    "Quantitative Research",

    # Humanities Skills
    "Historical Analysis",
    "Literary Criticism",
    "Philosophical Reasoning",

    # Business Skills
    "Market Research",
    "Entrepreneurship",

    # Health and Medicine Skills
    "Clinical Skills",
    "Public Health",

    # Arts and Design Skills
    "Artistic Skills",
    "Design Principles",

    # Engineering Skills
    "Technical Drawing",
    "Systems Design",

    # Language Skills
    "Language Proficiency",

    # Information Technology Skills
    "Cybersecurity Fundamentals",
    "Network Management",
    "Cloud Computing Skills"
    }

# Example usage
if __name__ == "__main__":
    email = input("Enter the email of the user: ")
    get_user_resume(email)
```
