```python
import os
from pdfminer.high_level import extract_text
from docx import Document

# Text extraction functions
def pdf_to_text(file_path):
    text = extract_text(file_path)
    return text

def doc_to_text(file_path):
    doc = Document(file_path)
    text = '\n'.join([paragraph.text for paragraph in doc.paragraphs])
    return text

# Main code
input_type = input("Enter 'file' for uploading a resume file in pdf or docx: ").strip().lower()

if input_type == 'file':
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]
    print(f"File '{file_name}' uploaded successfully.")

    if file_name.lower().endswith('.pdf'):
        text = pdf_to_text(file_name)
    elif file_name.lower().endswith('.docx'):
        text = doc_to_text(file_name)
    else:
        print("Unsupported file format. Please upload a PDF or DOCX file.")
        text = None

    if text:
        print("\nExtracted Text:")
        print(text)

        # Save the extracted text to a file
        output_file_path = "extracted_text.txt"
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(text)
        print(f"\nExtracted text saved to '{output_file_path}'.")

else:
    print("Invalid input type. Please enter 'file'")
```

```python
import spacy
import spacy_transformers
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# Ensure necessary NLTK resources are downloaded
nltk.download('stopwords')
nltk.download('punkt')

# Load the transformer-based SpaCy model
nlp = spacy.load("en_core_web_trf")

# Preprocessing function
def preprocess_text(text):
    # Tokenization
    sentences = sent_tokenize(text)
    words = word_tokenize(text)

    # Normalization
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation

    # Stop Words Removal
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Lemmatization
    lemmatized_words = []
    for word in filtered_words:
        doc = nlp(word)
        lemmatized_words.append(doc[0].lemma_)  # Use lemmatization

    # Combine all preprocessed text into a single string
    preprocessed_text = f"Sentences:\n{sentences}\n\nFiltered Words:\n{filtered_words}\n\nLemmatized Words:\n{lemmatized_words}"

    return preprocessed_text

# Read the extracted text from the file
extracted_text_path = "/content/extracted_text.txt"

with open(extracted_text_path, 'r', encoding='utf-8') as file:
    saved_text = file.read()

# Call the preprocessing function with the saved text
preprocessed_text = preprocess_text(saved_text)

# Print the preprocessed text
print("\nPreprocessed Text:")
print(preprocessed_text)

# Store the preprocessed text in a single text file
output_preprocessed_path = "/content/preprocessed_text.txt"

with open(output_preprocessed_path, 'w', encoding='utf-8') as file:
    file.write(preprocessed_text)

print(f"\nPreprocessed text saved to '{output_preprocessed_path}'.")
```
